How reliable are the [cm] interpreted values of the Sensor Module?

#### Following Plots of the Sensor Readings at different distances (5..150cm)

##### Defined Range of 3 cm

```python
import numpy as np
import matplotlib.pyplot as plt
import csv

with open('3cm_Data.csv', 'r') as f:
    reader = csv.reader(f, delimiter=';')
    values = np.array([float(row[0]) for row in reader])

sorted_values = np.sort(values)
median_value = np.median(sorted_values)

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# First plot: median filter
axs[0].bar(range(len(sorted_values)), sorted_values, align='center', alpha=0.5)
axs[0].axhline(median_value, color='r', linestyle='--', label=f'Median: {median_value:.2f}')
axs[0].legend()
axs[0].set_xlabel('3cm Range Values')
axs[0].set_ylabel('Distance in [cm]')
axs[0].grid()
axs[0].set_title('Median Filter')

# Second plot: values over time
time_periods = range(len(values))
axs[1].plot(time_periods, values, label='Distance over time')
axs[1].legend()
axs[1].set_xlabel('Time Periods')
axs[1].set_ylabel('Distance in [cm]')
axs[1].grid()
axs[1].set_title('Measurements over the time')

plt.tight_layout()
plt.show()
```
