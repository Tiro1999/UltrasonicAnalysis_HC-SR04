How reliable are the [cm] interpreted values of the Sensor Module?

#### Following Plots of the Sensor Readings at different distances (5..150cm)

##### Defined Range of 3 cm

```python
import numpy as np
import matplotlib.pyplot as plt
import csv

with open('3cm_Data.csv', 'r') as f:
    reader = csv.reader(f, delimiter=';')
    values = np.array([float(row[0]) for row in reader])

sorted_values = np.sort(values)
median_value = np.median(sorted_values)

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# First plot: median filter
axs[0].bar(range(len(sorted_values)), sorted_values, align='center', alpha=0.5)
axs[0].axhline(median_value, color='r', linestyle='--', label=f'Median: {median_value:.2f}')
axs[0].legend()
axs[0].set_xlabel('3cm Range Values')
axs[0].set_ylabel('Distance in [cm]')
axs[0].grid()
axs[0].set_title('Median Filter')

# Second plot: values over time
time_periods = range(len(values))
axs[1].plot(time_periods, values, label='Distance over time')
axs[1].legend()
axs[1].set_xlabel('Time Periods')
axs[1].set_ylabel('Distance in [cm]')
axs[1].grid()
axs[1].set_title('Measurements over the time')

plt.tight_layout()
plt.show()
```

### Other information about the measurements

Measure against an angeled surface of 30¬∞ in a distance of 13cm returns measured values of around 27cm while producing an error with an value over 30m every second reading.

### Minimum and Maximum Ratings
I measured a Distance up to 3.97m and the sensor returned a value of 382cm. Pretty close but there was an interferance of around 156cm in the returned values. In this distance was the first door frame that seemed to have a slight reflection.
Therefore you can conclude that the sensor is not very precise in the adjusted direction and the values are easy to manipulate with objects in a wider "field of view".
But it¬¥s still possible to get to the ground-truth values by fillig up an array and inspect it for any anomalies like this.

In a distance of around 3cm the returned values are a bit overrated (look at the first plotted CSV File)
If you are getting closer that 3cm the values start "jumping" in a range of +/- 0.6cm.
At some point < 2 cm the good looking values are way to high for the real distance and this is paired with a non-zero error of the known beyond 30m values.
 

### Gaussian distribution test 

more details will follow in the future for now it¬¥s just the first try

```python
import numpy as np
import matplotlib.pyplot as plt
import csv

with open('5cm_Data.csv', 'r') as f:
    reader = csv.reader(f, delimiter=';')
    values = np.array([float(row[0]) for row in reader])

mean_value = np.mean(values)
std_value = np.std(values)

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# First plot: Gaussian distribution
x = np.linspace(min(values), max(values), 100)
y = 1/(std_value*np.sqrt(2*np.pi)) * np.exp(-(x-mean_value)**2/(2*std_value**2))
axs[0].plot(x, y)
axs[0].set_xlabel('Distance in [cm]')
axs[0].set_ylabel('Probability Density')
axs[0].grid()
axs[0].set_title('Gaussian Distribution')

# Second plot: values over time
time_periods = range(len(values))
axs[1].plot(time_periods, values, label='Distance 5cm')
axs[1].legend()
axs[1].set_xlabel('Time Periods')
axs[1].set_ylabel('Distance in [cm]')
axs[1].grid()
axs[1].set_title('Measurements over the time')

plt.tight_layout()
plt.show()
```
